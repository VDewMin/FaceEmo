Data Preprocessing

```python
import os
import numpy as np 
import matplotlib.pyplot as plt 
from tensorflow.keras.utils import to_categorical
```

```python
#paths
train_dir = '../data/train/'
test_dir = '../data/test'

#create a mapping from emotion name to number - label encoding
emotion_labels = {'angry': 0, 'disgust':1, 'fear':2, 'happy':3, 'neutral':4, 'sad': 5, 'surprise': 6}

print("Emotion to number mapping:")
for emotion, number in emotion_labels.items():
    print(f" {emotion } → {number}")
```

```python
def load_images(data_dir, emotion_labels):
    images = []
    labels = []

    #loop through each emotion folder
    for emotion, label in emotion_labels.items():
        emotion_path = os.path.join(data_dir, emotion)

        #loop through each image in that folder
        for img_file in os.listdir(emotion_path):
            img_path = os.path.join(emotion_path, img_file)

            #read the image as grayscale
            img = plt.imread(img_path)

            #normalize pixel values from 0-255 to 0-1
            img = img / 255.0

            #Reshape from (48,48) to (48,48,1) - adding channel dimension
            img = img.reshape(48, 48, 1)

            images.append(img)
            labels.append(label)

    #convert lists to numpy arrays
    images = np.array(images)
    labels = np.array(labels)

    #shuffle data 

    indices = np.random.permutation(len(images))
    images = images[indices]
    labels = labels[indices]

    return images, labels
```

```python
print("Loading training data...")
X_train, y_train = load_images(train_dir, emotion_labels)

print("Loading test data...")
X_test, y_test = load_images(test_dir, emotion_labels)

print(f"\nTraining images shape: {X_train.shape}")
print(f"Training labels shpae: {y_train.shape}")
print(f"Test images shape: {X_test.shape}")
print(f"Test labels shape: {y_test.shape}")
```

```python
#convert number labels to one-hot encoded format
#model's output layer expects this

y_train_encoded = to_categorical(y_train, num_classes=7)
y_test_encoded = to_categorical(y_test, num_classes=7)

print("Before encoding - label example:", y_train[0])
print("After encoding - label example:", y_train_encoded[0])
```

```python
#saving processed arrays
np.save('../data/X_train.npy', X_train)
np.save('../data/y_train.npy', y_train_encoded)
np.save('../data/X_test.npy', X_test)
np.save('../data/y_test.npy', y_test_encoded)

print("Preprocessed data saved successfully!")
```
