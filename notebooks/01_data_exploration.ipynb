Exploratory Data Analysis


```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python

train_dir = '../data/train'

emotions = os.listdir(train_dir)
print("Emotions found:", emotions)
print("Total number of emotions:", len(emotions))
```

```python

for emotion in emotions:
    emotion_path = os.path.join(train_dir, emotion)
    count = len(os.listdir(emotion_path))
    print(f"{emotion}: {count} images")
```

```python

fig, axes = plt.subplots(1, len(emotions), figsize=(20, 4))

for i, emotion in enumerate(emotions):
    emotion_path = os.path.join(train_dir, emotion)

    
    img_path = os.path.join(emotion_path, os.listdir(emotion_path)[0])

     
    img = plt.imread(img_path)

    axes[i].imshow(img, cmap='gray')
    axes[i].set_title(emotion)
    axes[i].axis('off')

plt.suptitle('Sample Images for Emotion', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
sample_img_path = os.path.join(train_dir, emotions[0], os.listdir(os.path.join(train_dir, emotions[0]))[0])
sample_img = plt.imread(sample_img_path)
print("Images shape:", sample_img.shape)
print("Images data type:", sample_img.dtype)
print("Min pixel value:", sample_img.min())
print("Max pixel value:", sample_img.max())
```
